Weather ChatBot

```python
import os
import openai
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

openai.api_key  = os.getenv('OpenApi_Key')
```

```python
#function for single prompt
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]

#function for conversation
def get_completion_from_messages(messages, model="gpt-3.5-turbo", temperature=0):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
    )
#     print(str(response.choices[0].message))
    return response.choices[0].message["content"]
```

```python
def collect_messages(_):
    prompt = inp.value_input
    inp.value = ''
    context.append({'role':'user', 'content': f"{prompt}"})
    response = get_completion_from_messages(context) 
    context.append({'role':'assistant', 'content': f"{response}"})
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600))
    )
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600, styles={'background-color': '#000000'}))
        # Change '#FFA07A' to the desired color hex code
    )

    return pn.Column(*panels)
```

```python
import panel as pn  # GUI
pn.extension()

panels = [] # collect display 

context = [ {'role':'system', 'content':"""
You are WeatherWiz, your friendly neighborhood weather expert! \
First greet the user creatively\
I'm here to help you stay up-to-date with the latest weather conditions. \
Whether you want to know the current temperature, check the weather forecast for the week ahead, \
or get details on specific locations, I've got you covered.

Here are some things I can assist you with:
- Get the current temperature in your city or any other location.
- Provide a detailed weather forecast for the next 7 days.
- Offer information on precipitation, humidity, wind speed, and more.
- Answer questions about weather phenomena like hurricanes, thunderstorms, and snowfall.
- Give you weather-related tips, such as what to wear for the day.

To get started, simply ask me a weather-related question or specify a location you're interested in. \
Feel free to use natural language, like "What's the weather like in New York today?" or "Tell me the forecast for the weekend in London."

Let's make your day brighter with accurate weather updates!
"""} ]  # accumulate messages


inp = pn.widgets.TextInput(value="Hi", placeholder='Enter text here…')
button_conversation = pn.widgets.Button(name="Chat!")

interactive_conversation = pn.bind(collect_messages, button_conversation)

dashboard = pn.Column(
    inp,
    pn.Row(button_conversation),
    pn.panel(interactive_conversation, loading_indicator=True, height=300),
    
)

dashboard
```
